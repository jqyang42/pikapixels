```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df_census_initial = pd.read_csv("./data_sets/NC_censusData.csv")
df_sites = pd.read_csv("./data_sets/NC_testSites.csv")
```

```python
df_census_initial.head()
```

```python
df_census = df_census_initial[(df_census_initial.YEAR==4) & (df_census_initial.AGEGRP==0)]
df_pop = df_census[['CTYNAME', 'TOT_POP']]
df_pop = df_pop.sort_values(by=['CTYNAME'])
df_pop.CTYNAME = df_pop.CTYNAME.map(lambda x: x.rstrip(' County'))
df_pop
```

```python
df_sites.head()
```

```python
df_sites_nc = df_sites[df_sites.State == 'NC']
df_sites_nc.county.isnull().sum()
```

```python
dict_site = {}
for c in df_sites_nc.county:
    if c in dict_site:
    
        dict_site[c] = dict_site[c] + 1
    else:
        dict_site[c] = 1

df_site_count = pd.DataFrame.from_dict({"County":list(dict_site.keys()), "Site Count":list(dict_site.values())})
df_site_count = df_site_count.sort_values(by=['County'])
df_site_count
```
